```python
#wanted to make llm component which can be used in future

import random

class naklillm :
  def __init__(self):
    print("llm created")

  def predict(self, prompt):

    response_list = [
        'Delhi is the capital of india',
        'ipl is cricket',
        'i love ai'
    ]
    return {'response' : random.choice(response_list)}

llm = naklillm()   #llm class invoke
llm.predict('what is the capital of india') #it is random
```

```python
class NakliPromptTemplate :
  def __init__(self, template , input_variables):
    self.template = template
    self.input_variables = input_variables
  def format(self, input_dict):
    return self.template.format(**input_dict)

template = NakliPromptTemplate(
    template = 'write a {length} poem about {topic}',
    input_variables = ['length','topic']
)

prompt = template.format({'topic' : 'india', 'length' : 'short'})
```

```python
llm = naklillm()
```

```python
llm.predict(prompt)
```

Now creating LLM CHAIN , to join prompt template and llm , to make ease for ai engineers

```python
class nakliLLMchain :
  def __init__(self, llm, prompt):
    self.llm = llm
    self.prompt = prompt
  def run(self, input_dict):
    final_prompt = self.prompt.format(input_dict)
    result = self.llm.predict(final_prompt)

    return result['response']

template = NakliPromptTemplate(
    template = 'write a {length} poem about {topic}',
    input_variables = ['length','topic']
)

llm = naklillm()

chain = nakliLLMchain(llm , template)

chain.run({'length':'short' , 'topic':'naman'})
```

```python
from abc import ABC, abstractmethod
import random
class Runnable(ABC):
   #its a abstract class as it inherits from ABC

   @abstractmethod
   def invoke(input_data):
    pass
    #means invoke is abstract method , you have to call again and make function complete in your child class
```

```python
class naklillm(Runnable) :#child of Runnable class
  def __init__(self):
    print("llm created")

  def invoke(self, prompt):    #have to make this function as when i call naklillm , throws error as didnt find abstract class defined
  #forcing llm component to follow this structure
    response_list = [
        'Delhi is the capital of india',
        'ipl is cricket',
        'i love ai'
    ]
    return {'response' : random.choice(response_list)}

  def predict(self, prompt):

    response_list = [
        'Delhi is the capital of india',
        'ipl is cricket',
        'i love ai'
    ]
    return {'response' : random.choice(response_list)}   #cn attach warning with predict as in future predict thing going not to be there and invoke will be

class NakliPromptTemplate(Runnable) : #similarly for this also , child of runnable class

  def __init__(self, template , input_variables):
    self.template = template
    self.input_variables = input_variables

  #have to make same invoke function here also coz chill of abstract class
  def invoke(self, input_dict):
    return self.template.format(**input_dict)

  def format(self, input_dict):
    return self.template.format(**input_dict)

#NOW BOTH ARE RUNNABLES

#convert both the classes into Runnables
#and all the runnables must have some common methods , one of the most imps in INVOKE
#make sure through oops (abstract method)
# abstract class , all component classes inherits from runnable classes , auto all classes become runnable
```

NOW COMPOSING INTO CHAINS

```python
class nakliStringOutputParser(Runnable):
  def __init__(self):
    pass

  def invoke(self, input_data):
    return input_data['response']
```

```python
# NEED TO MAKE 1 CLASS, WHICH HELP TO CHAIN ONE OR MORE COMPONENTS

class RunnableChain(Runnable):#also abstract of runnable class

  def __init__(self, runnable_list):
    self.runnable_list = runnable_list

  def invoke(self, input_data):

    #traverse fron runnable_list passed when calling
    for runnable in self.runnable_list:
      input_data = runnable.invoke(input_data)

    return input_data  #return the last output


#CREATING OBJECTS TO PASS IN CHAIN
template = NakliPromptTemplate(
    template = 'write a {length} poem about {topic}',
    input_variables = ['length','topic']
)

llm = naklillm()

parser = nakliStringOutputParser()

#CALLING RunnableChain
chain = RunnableChain([template, llm, parser]) #list passed in runnable_list
#for ex : template.invoke(input_data)    will pass input_data
#and llm.invoke(input_data)   here input_data is output of previous


chain.invoke({'length' : 'long' , 'topic' : 'india'} )
```

TOOLKIT : 
its a collection of multiple related tools which serves a common purpose -
packaged together for convinience and reusability 

In langchain toolkit are : GoogleDriveToolKit having -> 
                                                    GoogleDriveCreateFileTool
                                                    GoogleDriveSearchTool
                                                    GoogleDriverReadFileTool

```python
from langchain_community.tools import tool

#custom tools -> have to make more than one related tools
@tool
def add(a : int, b : int) -> int :
    """Add two numbers"""
    return a + b

@tool 
def multiply(a : int, b : int) -> int :
    """Multiply two numbers"""
    return a*b
```

```python
class MathTooLKit : #joh bhi toolkit banani hai uske lie ek class banate hai
    def get_tools(self):
        return [add,multiply]
```

```python
toolkit = MathTooLKit() #creating object 
tools = toolkit.get_tools()    #doing this we get all tools we defined
####
#and can access by running loop
for tool in tools : 
    print(tool.name , tool.description)
```

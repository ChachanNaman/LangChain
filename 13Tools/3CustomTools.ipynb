```python

from langchain_core.tools import tool
from langchain_core import tools
```

```python
#Step-1 create a function 
def multiply(a,b):
    """Multiply two numbers"""  #provide doc string which makes llm understand what function doing
    return a*b
```

```python
#step-2 add type hints
def multiply(a: int, b : int) -> int :  #recomendded to do 
    """Multiply two numbers"""   
    return a*b
```

```python
#step-3 add tool decorator 

#iss function ko ek special func banadeta hai soo llm iss func ke sath communicate kar sakta hai
@tool
def multiply(a: int, b : int) -> int : 
    """Multiply two numbers"""   
    return a*b
```

```python
#since it tool now , its runnable now
result = multiply.invoke({"a" : 3, "b" : 4})
```
